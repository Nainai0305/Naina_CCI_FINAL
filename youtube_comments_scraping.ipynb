```python
import os
from googleapiclient.discovery import build

api_key = ""

youtube = build('youtube', 'v3', developerKey=api_key)

playlist_id = 'PLBVNJo7nhINSjBZdNezW15PzOTCc-10m9'

videos = []

nextPageToken = None
while True:
    pl_request = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=playlist_id,
        maxResults=50,
        pageToken=nextPageToken
    )

    pl_response = pl_request.execute()

    vid_ids = []
    for item in pl_response['items']:
        vid_ids.append(item['contentDetails']['videoId'])

    vid_request = youtube.videos().list(
        part="snippet,statistics",  # Include 'snippet' to get the title and publishing date
        id=','.join(vid_ids)
    )

    vid_response = vid_request.execute()

    for item in vid_response['items']:
        vid_views = item['statistics']['viewCount']
        vid_title = item['snippet']['title']  # Get the video title
        vid_published_at = item['snippet']['publishedAt']  # Get the publishing date

        vid_id = item['id']
        yt_link = f'https://youtu.be/{vid_id}'

        videos.append(
            {
                'title': vid_title,
                'views': int(vid_views),
                'url': yt_link,
                'published_at': vid_published_at
            }
        )

    nextPageToken = pl_response.get('nextPageToken')

    if not nextPageToken:
        break

videos.sort(key=lambda vid: vid['views'], reverse=True)

for video in videos[:163]:
    print(f"{video['title']} - {video['url']} - {video['views']} views - Published on {video['published_at']}")
```

```python
import pandas as pd  # Import Pandas for DataFrame manipulation

# Turn the videos list into a DataFrame
df = pd.DataFrame(videos, columns=['title', 'url', 'views', 'published_at'])

# Sort by views if you want to see the most-viewed videos first
df = df.sort_values(by='views', ascending=False)

# Show the top 10 videos by views
print(df.head(10))

# Optionally, save the DataFrame to a CSV file
csv_file_path = "youtube_videos_dataframe.csv"
df.to_csv(csv_file_path, index=False, encoding='utf-8')
print(f"DataFrame has been saved to {csv_file_path}")
```
